It is clear that exist strong correlation betewen distance of a neighbourhood to a nearestfrom pharmacy and the necessity in a new closer pharmacy. Subsequently, I will find the distance between every neighbourhood and a nearest pharmacy.  

```python
# Using geopy to calculate distance in kilometers beetwen every Neighbourhood and every Pharmacy
import geopy.distance
rr = []
for r in range(len(result_a)):
    for j in range(len(dataframe)):    
        coords_1 = (result_a.iloc[r]['Latitude'], result_a.iloc[r]['Longitude'])
        coords_2 = (dataframe.iloc[j]['location.lat'], dataframe.iloc[j]['location.lng'])
        dist = geopy.distance.distance(coords_1, coords_2).km
        rr.append([result_a.iloc[r]['Neighbourhood Name'], dist])
```

```python
# Create DataFrame and place in it the name, the latitude and the longitude for Neighbourhoods and Pharmacies as well as distance beetwen them
df = pd.DataFrame(rr)
headers = ["Neighbourhood", "distance"]
df.columns = headers
df.head ()
```

I received the DataFrame that contains distance from every neighbourhood to every pharmacy. The next step is to find the minimum distance, convert it to meters, and cut off neighbourhoods with distance to a nearest pharmacy less than 2500 meters.

```python
# Finding the minimum distance for all the Neighbourhoods and sort by distance
df_cc = df.groupby('Neighbourhood')['distance'].min().to_frame().reset_index().sort_values(by='distance')
# Convert distance to metres
df_cc['distance'] *= 1000
df_cc['distance'] = df_cc['distance'].astype(int) 
# Cut Neighbourhoods with distance to a pharmacy less than 2500 metres
df_cc = df_cc[df_cc['distance'] > 2500]
```

To continue the research I need to add more neibourhood data into the Frame. 

```python
result_fn = pd.merge(left=df_cc, right=result_a, how='left', left_on='Neighbourhood', right_on='Neighbourhood Name')
result_fn = result_fn.drop (['Neighbourhood Number', 
                             'Neighbourhood Name'], axis = 1)
result_fn.head ()
```

I received the DataFrame that contains all the necessary information about neighbourhoods that are away from the nearest pharmacy. 

```python
# create map of Edmonton using latitude and longitude values
map_Ed = folium.Map(location=[latitude, longitude], zoom_start=10)
# add markers to map
for lat, lng, borough, neighborhood, inh in zip(result_fn['Latitude'], result_fn['Longitude'], result_fn['distance'], result_fn['Neighbourhood'], result_fn['Inhabitants']):
    label = '{}, {}'.format(neighborhood, inh)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=inh/500,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_Ed)  
map_Ed
```

This neighbourhoods distributed in the north, the south-east and the west of Edmonton mostly. The sum of inhabatants of this neyghbourhoods is ..... From the data it can be seen that the inhabitants living close to a pharmacy is for the total number of pharmacies . It is clear, that the same proportion is used for those who live away. 

```python
# Find the desired quantity of additional pharmacies
k = (len(dataframe) * result_fn['Inhabitants'].sum ())/(result_a['Inhabitants'].sum () - result_fn['Inhabitants'].sum ())
k
```

As we can see it should be opened 11 (10.875 rounded to nearest) new pharmacies. 
The next problem to be solved is coordinates of the new pharmacies. The good idea in that is to use K-Means Clustering. 
I will use the initialization method of the centroids k-means++ and the number of time the k-means algorithm will be run with different centroid seeds is 10.

```python
k_means = KMeans(init = "k-means++", n_clusters = 11, n_init = 10)
```

```python
# convert the Pandas data frame to a Numpy array
X = result_fn[['Latitude', 'Longitude']].values  
```

```python
k_means.fit(X)
```

```python
# Get the coordinates of the cluster centers using KMeans
k_means_cluster_centers = k_means.cluster_centers_
```

```python
# Create DataFrame and place in it the latitude and the longitude for new Pharmacies
df_nph = pd.DataFrame(k_means_cluster_centers)
headers = ["Latitude", "Longitude"]
df_nph.columns = headers
df_nph.head ()
```

```python
# create map of Edmonton using latitude and longitude values
map_Ednph = folium.Map(location=[latitude, longitude], zoom_start=10)
# add markers to map
for lat, lng in zip(df_nph['Latitude'], df_nph['Longitude']):
    folium.CircleMarker(
        [lat, lng],
        radius=10,
        color='blue',
        fill=True,
        fill_color='#1ed2e3',
        fill_opacity=0.7,
        parse_html=False).add_to(map_Ednph)  
map_Ednph
```
